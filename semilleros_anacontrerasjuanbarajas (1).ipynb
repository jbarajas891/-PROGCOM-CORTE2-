•	Semillero en Energía Nuclear

Línea: Producción y estudio de plasma

Proyecto: Aplicación de técnicas de diagnostico a un plasma generado a partir de fuego excitado por microondas

Se reconoce que el plasma es el estado de la materia que se encuentra en mayor cantidad en el universo, cuenta con propiedades únicas las cuales benefician a industrias, mediante este proyecto se busca la utilización de técnicas de diagnostico en un plasma generado a partir de fuego excitado por microondas, la cámara es diseñada teniendo en cuenta que esta debe satisfaces una gran  concentración electromagnética, luego de esto, se procede a determinar las características del plasma y finalmente identificar el comportamiento del mismo dentro de la cámara gracias a electroimanes.

•	Opalo reseach group- Semillero de Investigación OPALO - Fase III

Línea: “Desarrollo de modelos de optimización, simulación, big data analytics y manufactura inteligente” (tomado de https://www.opaloresearchgroup.com/)

Proyecto: Solución al problema de ruteo de vehículos de flota heterogénea en transbordo lateral mediante la metaheurística de algoritmo genético 
En Colombia, existen problemas en cuanto al transporte, especialmente, las flotas de vehículos heterogéneas debido a que son más usuales y se encuentran en mayor proporción que las homogéneas. Mediante este proyecto se busca el diseño de un modelo matemático y su correspondiente solución real y factible a través de un algoritmo  para el VRP de flota heterogénea con transbordo lateral, por supuesto, antes de salir al mercado se debe hacer una validación del algoritmo diseñado y si es el caso, realizar ajustes para que este funcione correctamente en su totalidad, finalmente, se haría una documentación de los resultados obtenidos.

•	Semillero SEBIUNAB-Programa de ingeniería biomédica 

Línea: Monitoreo variables fisiológicas 

Proyecto: Sistema de sincronía entre equipos de monitoreo de variables fisiológicas 
Al momento de tomar decisiones, el sistema nervioso juega un papel importante ya que influye en nuestras emociones y sensaciones, que puede considerarse como ’una mala jugada ’ para el humano. Mediante este proyecto se propone el monitoreo de variables fisiológicas a través de equipos biomédicos mediante el software de MATLAB, luego de esto, se procederá a analizar la sincronización de bases de datos obtenidos, esto con el objetivo de analizar y caracterizar las variables fisiológicas escogidas.

•	SISAS semillero de investigacion en aire, suelo y sostenibilidad- GAIA grupo ambiental de investigacion aplicada

Línea: Calidad de aire y cambio climático

Proyecto: Evaluacion del ruido ambiental en el campus de la Universidad de Santander con un módulo electrónico de bajo costo en tiempo real
Se creo un dispositivo para medir la calidad del sonido de la Universidad, teniendo en cuenta que, es un espacio con muchas zonas verdes. Fue ubicado en tres puntos clave de la universiad ya que estos se consideraban ‘generadores de ruido’, es un dispositivo de bajo costo, sin embargo, se busca hacer un modelo mas pulido y estético del mismo para que pueda salir a la venta y tenga una gran acogida ya que es muy útil, a su vez, se van a generar mejoras en la aplicación para que el usuario pueda visualizar de manera organizada los resultados de los niveles de presión sonora.

•	Semillero medicina regenerativa- ingeniería biomédica

Línea: medicina regenerativa

Actualmente, las células madre mesenquimales han tomado un gran lugar en la medicina regenerativa debido a su capacidad de autorrenovarse, principalmente, han sido utilizadas en la regeneración de tejidos. En algunos casos, las condiciones ambientales influyen en aumentar la posibilidad de muerte celular, por ello, el medio de cultivo tiene que ser un medio ideal que permita un ciclo de vida de las células para que alcancen a cumplir los objetivos para los que están siendo implementadas. La finalidad de este proyecto es evaluar el efecto de las condiciones hipóxicas y medio no xenogénico en la expansión de MSC, es decir, se plantea la elaboración de cultivos in vitro, lo que fomenta la expansión de células madre para que se puedan aplicar en medicina regenerativa.

•	FACARQ 

Línea: diseño arquitectónico

Poryecto: Recursos didácticos desde la arquitectura que identifican la flora nativa en el área metropolitana de Bucaramanga
Este proyecto reconoce que al momento de vender un apartamento o casa en Bucaramanga y su área metropolitana se ignora o no es importante el aspecto de la zona verde del conjunto, las empresas no fomentan la flora de la región, al contrario, traen flora de otras regiones ya que es mas económico. Mediante este proyecto se busca generar un paquete de recursos didácticos para la representación gráfica de la vegetación en los planos de arquitectura, a través de la identificación de la flora nativa de Bucaramanga y su área metropolitana. Se esperan gráficos de tres especies nativas, plantillas de dibujo para dibujo a mano alzada y video explicativo.

•	Semillero de investigación Hoshin- grupo de investigación SINERGIA

Línea: sector de servicios

Proyecto: Industria 4.0 y economía circular en el sector servicios
Actualmente, la sostenibilidad es un aspecto en el que está enfocado cualquier sector económico, social y ambiental. Mediante este proyecto se propone determinar estrategias que las empresas colombianas del sector servicios deben tener en cuenta para que implementen la industrial 4.0 y el modelo de economía circular teniendo en cuenta el desarrollo sostenible. Se busca que los recursos sean utilizados en su totalidad y no sean desaprovechados, esto fomenta la creación de empresas y por ende, creación de empleo.
