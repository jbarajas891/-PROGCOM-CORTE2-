```python
import random
import math
import os
## editado por juan pablo barajas un capitan
#X es lo maximo = 1
#O minimo = -1, si se colocan negativos se empieza por cero en la parte inferior derecha, y el orden sera de 0,-1...-9 de
#de derecha a ixquierda

class TicTacToe:
    def __init__(self):
        self.board = ['-' for _ in range(9)]
        if random.randint(0, 1) == 1:
            self.jugadorhumano = 'X'
            self.maquina = "O"
        else:
            self.jugadorhumano = "O"
            self.maquina = "X"

    def show_board(self):
        print("")
        for i in range(3):
            print("  ",self.board[0+(i*3)]," | ",self.board[1+(i*3)]," | ",self.board[2+(i*3)])
            print("")
            
    def is_board_filled(self,state):
        return not "-" in state

    def is_player_win(self,state,player):
        if state[0]==state[1]==state[2] == player: return True
        if state[3]==state[4]==state[5] == player: return True
        if state[6]==state[7]==state[8] == player: return True
        if state[0]==state[3]==state[6] == player: return True
        if state[1]==state[4]==state[7] == player: return True
        if state[2]==state[5]==state[8] == player: return True
        if state[0]==state[4]==state[8] == player: return True
        if state[2]==state[4]==state[6] == player: return True

        return False

    def checkWinner(self):
        if self.is_player_win(self.board,self. jugadorhumano):
            os.system("cls")
            print(f"   el jugador  {self. jugadorhumano} ha ganado eljuego!")
            return True
            
        if self.is_player_win(self.board,self.maquina):
            os.system("cls")
            print(f"   el jugador  {self.maquina} ha ganado eljuego!")
            return True

        # comprobar si el juego es empate o no
        if self.is_board_filled(self.board):
            os.system("cls")
            print("   empate del partido es imposible ganar a la IA!")
            return True
        return False

    def start(self):
        bot = ComputerPlayer(self.maquina)
        human = jugadorhumano(self.jugadorhumano)
        while True:
            os.system("cls")
            print(f"   es turno de la {self.jugadorhumano} :)")
            self.show_board()
            
            #el humano
            square = human.human_move(self.board)
            self.board[square] = self.jugadorhumano
            if self.checkWinner():
                break
            
            #la maquina o bot
            square = bot.machine_move(self.board)
            self.board[square] = self.maquina
            if self.checkWinner():
                break

        # mostrando la vista final del tablero
        print()
        self.show_board()

class jugadorhumano:
    def __init__(self,letter):
        self.letter = letter
    
    def human_move(self,state):
        # tomando la entrada del usuario
        while True:
            square =  int(input("Elija un lugar entre las nueve posiciones posibles empezando desde la superior izquierda hasta la derecha (1-9): "))
            print()
            if state[square-1] == "-":
                break
        return square-1

class ComputerPlayer(TicTacToe):
    def __init__(self,letter):
        self.maquina = letter
        self.jugadorhumano = "X" if letter == "O" else "O"

    def players(self,state):
        n = len(state)
        x = 0
        o = 0
        for i in range(9):
            if(state[i] == "X"):
                x = x+1
            if(state[i] == "O"):
                o = o+1
        
        if(self.jugadorhumano == "X"):
            return "X" if x==o else "O"
        if(self.jugadorhumano == "O"):
            return "O" if x==o else "X"
    
    def actions(self,state):
        return [i for i, x in enumerate(state) if x == "-"]
    
    def result(self,state,action):
        newState = state.copy()
        player = self.players(state)
        newState[action] = player
        return newState
    
    def terminal(self,state):
        if(self.is_player_win(state,"X")):
            return True
        if(self.is_player_win(state,"O")):
            return True
        return False

    def minimax(self, state, player):
        max_player = self.jugadorhumano  # tú mismo
        other_player = 'O' if player == 'X' else 'X'

        # primero se comprueba si el movimiento anterior es un ganador
        if self.terminal(state):
            return {'posición': None, 'puntaje': 1 * (len(self.actions(state)) + 1) if other_player == max_player else -1 * (
                        len(self.actions(state)) + 1)}
        elif self.is_board_filled(state):
            return {'posición': None, 'puntaje': 0}

        if player == max_player:
            best = {'posición': None, 'puntaje': -math.inf}  # cada puntaje debe maximizar
        else:
            best = {'posición': None, 'puntaje': math.inf}  # cada puntaje debe minimizar
        for possible_move in self.actions(state):
            newState = self.result(state,possible_move)
            sim_puntaje = self.minimax(newState, other_player)  # simular un juego después de hacer ese movimiento

            sim_puntaje['posición'] = possible_move  # esto representa el movimiento óptimo próximo movimiento

            if player == max_player:  
                if sim_puntaje['puntaje'] > best['puntaje']:
                    best = sim_puntaje
            else:
                if sim_puntaje['puntaje'] < best['puntaje']:
                    best = sim_puntaje
        return best

    def machine_move(self,state):
        square = self.minimax(state,self.maquina)['posición']
        return square

# comieza el partido
tic_tac_toe = TicTacToe()
tic_tac_toe.start()
```
